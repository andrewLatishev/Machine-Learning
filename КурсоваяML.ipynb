### Спарсим из столбца point значения lat и long, сделаем их отдельными полями. Point за ненадобностью удалим.

```python
import ast
df_dtp["point"] = df_dtp["point"].apply(lambda point: ast.literal_eval(f'{point}') if not pd.isnull(point) else point)

latitude = df_dtp['point'].apply(lambda point: point.get('lat') if isinstance(point, dict) else None)
longitude = df_dtp['point'].apply(lambda point: point.get('long') if isinstance(point, dict) else None)

df_dtp['lat'] = latitude
df_dtp['long'] = longitude

df_dtp = df_dtp.drop(columns=['point'])
```

```python
Y_balanced.value_counts()
```

```python
standard_scaler = StandardScaler()

X_standard = standard_scaler.fit_transform(X_balanced)
X = pd.DataFrame(X_standard, columns = X_balanced.columns, index = X_balanced.index)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y_balanced, test_size=0.33, random_state=42, stratify=Y_balanced)

print(f'Размер выборок: \nX: Train: {X_train.shape} Test: {X_test.shape} \nY: Train: {y_train.shape} Test: {y_test.shape}')
```

```python
y_train_umap = y_train.copy()
y_test_umap = y_test.copy()
```

```python
# Снижение размерности
from umap.umap_ import nearest_neighbors
import umap
import umap.plot
```

```python
# для ускорения обучения UMAP
mnist_knn = nearest_neighbors(X_train,
                              n_neighbors=250,
                              metric='braycurtis',
                              metric_kwds=None,
                              angular=False,
                              random_state=None)

# применение UMAP
data_umap = umap.UMAP(n_neighbors=100, metric='braycurtis', precomputed_knn=mnist_knn).fit(X_train, y_train_umap)
X_umap = data_umap.transform(X_train)
X_umap_test = data_umap.transform(X_test)
```

```python
IndianRed =	'#CD5C5C'
Olive =	'#808000'
PaleGoldenrod =	'#EEE8AA'
SteelBlue =	'#4682B4'

colors = [IndianRed, Olive, PaleGoldenrod, SteelBlue]

palette = sns.color_palette(colors)
sns.palplot(palette)

plt.title('Цвета для визуализаций')
plt.axis('off')
plt.show()
```

```python
def draw_umap(data_x, data_y):
    '''
    Построение графиков в двумерном пространстве
    '''

    all_data_umap = pd.DataFrame(data_x)
    all_data_umap.columns = ['Параметр_1', 'Параметр_2']
    all_data_umap['status'] = data_y
    all_data_umap['status_color'] = [palette[0] if x == 0 else palette[2] for x in all_data_umap['status']]
    all_data_umap.plot.scatter(x='Параметр_1', y='Параметр_2', c='status_color', s=5, alpha=0.5)
    plt.title("Отображение данных в двумерном пространстве", pad=15)
    plt.show()
    return all_data_umap
```

### Логистическая регрессия

```python
from sklearn.linear_model import LogisticRegression
```

```python
def objective(trial):
    C = trial.suggest_categorical('C', [0.1, 0.2, 0.5, 1, 2, 5, 10, 50, 100, 1000])

    lr_model = LogisticRegression(C=C, n_jobs=-1)

    scores = cross_val_score(lr_model, X_umap, y_train_umap, cv=5, scoring='accuracy')

    return scores.mean()

study = optuna.create_study(direction='maximize', sampler=GridSampler({'C': [0.1, 0.2, 0.5, 1, 2, 5, 10, 50, 100, 1000]}))

study.optimize(objective, n_trials=10)

best_params_lr = study.best_params
print(f"Лучшие параметры: {best_params_lr}")
```

```python
lr_model = LogisticRegression(C=1000, n_jobs=-1)

lr_model.fit(X_umap, y_train_umap)

lr_model_preds = lr_model.predict(X_umap_test)
lr_model_prob = lr_model.predict_proba(X_umap_test)
```

```python
print("Метрики для логистической регрессии")

lr_accuracy = accuracy_score(y_test_umap, lr_model_preds)
print(f'Accuracy: {lr_accuracy}')

lr_logloss = log_loss(y_test_umap, lr_model_prob)
print(f'Logloss: {lr_logloss}')

lr_f1 = f1_score(y_test_umap, lr_model_preds)
print(f'F1: {lr_f1}')
```

```python
fpr, tpr, thresholds = roc_curve(y_test_umap, lr_model_prob[:, 1])
roc_auc = roc_auc_score(y_test_umap, lr_model_prob[:, 1])

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f'ROC-AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Случайный выбор')
plt.xlabel('Ложноположительная оценка (False Positive Rate)')
plt.ylabel('Истинноположительная оценка (True Positive Rate)')
plt.title('ROC-кривая')
plt.legend()
plt.show()
```

```python
conf_matrix = confusion_matrix(y_test_umap, lr_model_preds)

plt.figure(figsize=(5, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Предсказанные')
plt.ylabel('Настоящие значения')
plt.title('Матрица ошибок для Логистической регрессии')
plt.show()
```

### Дерево решений

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
def objective(trial):
    params_grid = {
        'max_depth': trial.suggest_int('max_depth', 1, 20),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 10),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 20),
        'max_leaf_nodes': trial.suggest_int('max_leaf_nodes', 2, 20),
        'max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2']),
        'criterion': trial.suggest_categorical('criterion', ['gini', 'entropy', 'log_loss'])
    }

    dt_model = DecisionTreeClassifier(**params_grid)

    scores = cross_val_score(dt_model, X_umap, y_train_umap, cv=5, scoring="accuracy")

    return scores.mean()

study_dt = optuna.create_study(direction='maximize', sampler=RandomSampler(seed=42))

study_dt.optimize(objective, n_trials=10)

best_params_dt = study_dt.best_params
print(f"Лучшие параметры: {best_params_dt}")
```

```python
dt_model = DecisionTreeClassifier(**best_params_dt)

dt_model.fit(X_umap, y_train_umap)

dt_model_preds = dt_model.predict(X_umap_test)
dt_model_prob = dt_model.predict_proba(X_umap_test)
```

```python
print("Метрики для дерева решений")

dt_accuracy = accuracy_score(y_test_umap, dt_model_preds)
print(f'Accuracy: {dt_accuracy}')

dt_logloss = log_loss(y_test_umap, dt_model_prob)
print(f'Logloss: {dt_logloss}')

dt_f1 = f1_score(y_test_umap, dt_model_preds)
print(f'F1: {dt_f1}')
```

```python
fpr, tpr, thresholds = roc_curve(y_test_umap, dt_model_prob[:, 1])
roc_auc = roc_auc_score(y_test_umap, dt_model_prob[:, 1])

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f'ROC-AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Случайный выбор')
plt.xlabel('Ложноположительная оценка (False Positive Rate)')
plt.ylabel('Истинноположительная оценка (True Positive Rate)')
plt.title('ROC-кривая')
plt.legend()
plt.show()
```

### Случайный лес

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
def objective(trial):
    params_grid = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 100),
        'max_depth': trial.suggest_int('max_depth', 1, 30),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 10),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 4),
        'max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2'])
    }

    rf_model = RandomForestClassifier(**params_grid, n_jobs=-1)

    scores = cross_val_score(rf_model, X_umap, y_train_umap, cv=5, scoring="accuracy")

    return scores.mean()

study = optuna.create_study(direction='maximize', sampler=RandomSampler(seed=42))

study.optimize(objective, n_trials=10)

best_params_rf = study.best_params
print(f"Лучшие параметры: {best_params_rf}")
```
